```python
import urllib.request
import json

import tensorflow as tf
```

```python
# https://huggingface.co/datasets/wikitext/viewer/wikitext-103-v1/train
WIKI_URL_FMT = "https://datasets-server.huggingface.co/rows?dataset=wikitext&config=wikitext-103-v1&split=train&offset={}&limit=100"
PER_CALL_LIMIT = 100
WIKI_DATASET = 'export/raw_wiki_dataset'
```

```python
def process_row(row):
    paragraph = row['row']['text']
    sentences = []
    for sentence in paragraph.split('.'):
        sentence = sentence.strip()
        if len(sentence) > 0 and sentence != '\n':
            sentences.append(sentence)
    return sentences

def filter_row(sentences):
    return len(sentences) > 1
```

```python
def wiki_generate(wiki_content):
    def _gen():
        for row in wiki_content['rows']:
            row = process_row(row)
            if filter_row(row):
                yield row
    for row in _gen():
        yield row
```

```python
def retries(fn, num_retries):
    for i in range(num_retries):
        try:
            return fn()
        except Exception as e:
            print('problem: {}, retrying {}...'.format(e, i))
    return None

failed_calls = []
def gen():
    page = 0
    while True:
        offset = page * PER_CALL_LIMIT
        wiki_url = WIKI_URL_FMT.format(offset)
        print(wiki_url) 
        def _wiki_getter():
            return json.loads(urllib.request.urlopen(wiki_url).read())
        wiki_content = retries(_wiki_getter, 5)
        if wiki_content is None:
            failed_calls.append(wiki_url)
        elif len(wiki_content['rows']) == 0:
            break
        else:
            for row in wiki_generate(wiki_content):
                yield row
        page += 1
```

```python
ds = tf.data.Dataset.from_generator(gen, output_signature=(tf.TensorSpec(shape=(None), dtype=tf.string)))
tf.data.experimental.save(ds, WIKI_DATASET)
```
